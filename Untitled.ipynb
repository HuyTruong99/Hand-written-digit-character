#Loading - "MNIST Data Set" ( which is handwritten digits dataset)
## Containing Training samples = 60,000, testing samples= 10,000
###Tensorflow already contain MNIST data set which can be loaded using Keras ( api of tensorflow)

```python
mnist = tf.keras.datasets.mnist ##this is basically handwritten character based on 28x28 sized images of 0 to 9
```

After, loading the MNIST data, divide into train and test datasets

```python
##unpacking the dataset into train and test datasets
(x_train, y_train),(x_test,y_test) = mnist.load_data()
```

```python
# you might have noticed that , its gray image and all values varies from 0 to 255
# in order to normalize it 

x_train = tf.keras.utils.normalize (x_train, axis = 1)
x_test = tf.keras.utils.normalize(x_test, axis=1)
plt.imshow(x_train[0], cmap=plt.cm.binary)
```

###resizing image to make it suitable for apply Convolution operation 

```python
import numpy as np #pip install numpy
IMG_SIZE=28
x_trainr = np.array(x_train).reshape(-1, IMG_SIZE, IMG_SIZE,1) ###increasing one dimension for kernel = filter operation
x_testr = np.array(x_test).reshape(-1, IMG_SIZE, IMG_SIZE,1)  ### increasing one dimension for kernel operation
print("Training Samples dimension", x_trainr.shape)
print("Testing Samples dimension", x_testr.shape)
```

# creating a Deep Neural Network 
# Training on 60,000 samples of MNIST handwritten dataset

```python
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten, Conv2D, MaxPooling2D
```

```python
### Creating a neural network now
model = Sequential()

#### First convolution layer 0 1 2 3 (6000, 28, 28 ,1) 28-3+1 = 26 * 26
model.add(Conv2D(64,(3,3), input_shape = x_trainr.shape[1:])) ###only for  first convolution layer to mention input 
#layer size
model.add(Activation("relu"))###activation function to make it non-linear , <0 , remove these value , >0 
model.add(MaxPooling2D(pool_size=(2,2)))###Maxpooling single maximum value of 2x2, 


### Second convolution layer
model.add(Conv2D(64, (3,3))) ##2nd convolution layer
model.add(Activation("relu")) ##activation function
model.add(MaxPooling2D(pool_size=(2,2))) ##Maxpooling because we have it so the size is divided 2 

### Third convolution layer
model.add(Conv2D(64, (3,3)))# 24*24
model.add(Activation("relu"))
model.add(MaxPooling2D(pool_size=(2,2)))

### Fully Connected Layer #1
model.add (Flatten()) ### before using fully connected layer,need to be flatten so that 2D to 1D
model.add (Dense(64)) #neural network 
model.add (Activation("relu"))

###Fully Connected layer #2
model.add (Dense(32))
model.add(Activation("relu"))


### lastly fully connected layer,output must be equal to number of classes , 10 (0-9)
model.add (Dense(10)) ## this last dense must be equal to 10
model.add(Activation('softmax')) ###activation function is changed to softmax ( class probabilities)
##binary classification , one neuron in Dense layer,sigmoid
```

```python
model.summary()
```

```python
print("Total Training Samples = ", len(x_trainr))
```

```python
import tensorflow as tf
optimizer = tf.compat.v1.train.GradientDescentOptimizer(learning_rate=0.1)
model.compile(loss="sparse_categorical_crossentropy", optimizer="adam", metrics=['accuracy'])
```

```python
model.fit(x_trainr,y_train,epochs=5, validation_split = 0.3)##Training my model
```

# in keras , there are two ways to implement it is the function api and sequence api 

```python
### Evaluating on testing data set MNIT 
test_loss, test_acc = model.evaluate(x_testr, y_test)
print("Test loss on 10,000 test samples", test_loss)
print("Validation Accuracy on 10,000 test samples", test_acc)
```
